```python
#paddledetection的工作需要严格遵循一定的数据集结构和命名方式
import os
import shutil

def rename_folder(original_folder_path, new_folder_name):
    new_folder_path = os.path.join(os.path.dirname(original_folder_path), new_folder_name)

    # 创建新的目标文件夹
    os.mkdir(new_folder_path)

    # 移动文件夹中的文件到新的目标文件夹
    file_list = os.listdir(original_folder_path)
    for file_name in file_list:
        file_path = os.path.join(original_folder_path, file_name)
        shutil.move(file_path, new_folder_path)

    # 删除原始文件夹
    os.rmdir(original_folder_path)
```

```python
#生成预测结果文件
import os
import json
root_path='/home/aistudio/toolinfer2'
for path in os.listdir(root_path):
    if(path.endswith('json')):
        json_path=os.path.join(root_path,path) 
with open("/home/aistudio/helmet/test_name.txt",'r') as f:
    target_lines =f.readlines()
test_data = []
for l in target_lines :
    test_data.append(l.strip())
test_data[5][39:-4]
cnt=0
with open("/home/aistudio/helmet/test_name.txt","r") as f:
    target_lines = f.readlines()
for l in target_lines:
    test_data.append(l.strip()[37:-3])
with open(json_path,'r') as f:
    data = json.load(f)
i=0
with open("pred_result.txt","w") as final_output:
    while(i<len(data)):
        if data[i]['score']<0.35:
            i+=1
            continue
        else:
            line=test_data[data[i]['image_id']][38:-4]+' '+str(data[i]['score'])+' '+str(data[i]['bbox'][0])+' '+str(data[i]['bbox'][1])+' '+str(data[i]['bbox'][2]+data[i]['bbox'][0])+' '+str(data[i]['bbox'][3]+data[i]['bbox'][1])+' '+str(int(data[i]['category_id'])-1)
            final_output.write(line+'\n')
            i+=1
```
